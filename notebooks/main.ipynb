```python
import pandas as pd
import matplotlib.pyplot as  plt
import seaborn as sns

```

```python
car_history = pd.read_csv('../data/history/gen0_car0.csv')
```

```python
car_history.head()
```

```python
# scale the direction column values to visualize better
car_history['direction'] = car_history['direction'].apply(lambda c: c*10+50)

features = [f for f in list(car_history.columns)]
ticks = [i for i in range(len(car_history))]

fig, axs = plt.subplots(1, 1, figsize=(20,20))

for i, feature in enumerate(features):
    axs.tick_params(axis='x', colors='white', labelsize=30)
    axs.tick_params(axis='y', colors='white', labelsize=30)
    sns.lineplot(ax=axs, x=ticks, y=feature, label=feature, data=car_history)

plt.legend(fontsize=30)
fig.tight_layout()
```
